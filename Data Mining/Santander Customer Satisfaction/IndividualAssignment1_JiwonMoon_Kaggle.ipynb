# Preprocessing the Data Set

### 1. Deleting 0 Variance

```python
# Finding values with 0 variance

variance = 0
for col in df_train.columns:
    if df_train[col].var()==0:
        variance += 1

print(variance)
```

-> There are 34 columns with 0 variance

```python
# Deleting columns with 0 variance(34 columns)

df_train_del = df_train.copy()
df_test_del = df_test.copy()

variance = 0
for col in df_train_del.columns:
    if df_train_del[col].var()==0:
        variance += 1
        del df_train_del[col]
        del df_test_del[col]
      
df_train_del.info()
df_test_del.info()
```

### 2. Deleting Duplicated Columns

```python
import itertools
from itertools import combinations

combination = list(itertools.combinations(df_train_del.columns,2))
print(combination)

duplicate = []
for t1,t2 in combination: 
    if (t1 not in duplicate) & (t2 not in duplicate):
        if df_train_del[t1].equals(df_train_del[t2]):
            duplicate.append(t1)
df_train_del.drop(duplicate, axis=1, inplace=True)
df_test_del.drop(duplicate, axis=1, inplace=True)

df_train_del.info()
df_test_del.info()
df_train.info()
df_test.info()
```

### 3. Find Null Values

```python
# Finding the null value

na_train = df_train.isnull()
na_sum_train = na_train.sum()
na_sum_train = na_sum_train.sort_values(ascending=False)

na_sum_train
```

-> There are no null values

# EDA

## 1. Finding the Outlier

### 1-1. mean +- 3* stdev -> 99.7%

```python
num_over = 0
num_outlier = 100

columns = df_train.columns
for col in columns:
    var = df_train[col]
    var_mean = df_train[col].mean()
    var_std = df_train[col].std()

    lower_limit = var < var_mean - 3* var_std
    upper_limit = var > var_mean + 3* var_std

    num_outliers = lower_limit.sum() + upper_limit.sum()
    if num_outliers > num_outlier:
      num_over += 1
      print("  columns: {}".format(col))
      print("    -> num of outlier : {}\n".format(num_outliers))
      
print("number of columns with outliers: {}\n".format(num_over))
```

-> number of columns with outliers: 187

```python
# Column names that have outliers

col_outliers = []

for col in columns:
    var = df_train[col]
    var_mean = df_train[col].mean()
    var_std = df_train[col].std()

    lower_bound_idx = var < var_mean - 3* var_std
    upper_bound_idx = var > var_mean + 3* var_std

    
    num_outliers = lower_bound_idx.sum() + upper_bound_idx.sum()
    if num_outliers > num_outlier:
      num_over += 1
      col_outliers.append(format(col))

col_outliers
```

### 1-2. Replacing the Outliers

```python
# Create a new DataFrame with outliers

df_col_outliers = df_train[df_train.columns[df_train.columns.isin(col_outliers)]]
df_col_outliers
```

## 2. Finding Important Features

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score

y = df_train.TARGET
X = df_train.drop(['TARGET', 'ID'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

train_cnt = y_train.count()
test_cnt = y_test.count()
print('Train_set Shape:{0}, Test_set Shape:{1}'.format(X_train.shape , X_test.shape))

print('Satisfaction Distribution of Train_set Variables')
print(y_train.value_counts()/train_cnt)
print('Satisfaction Distribution of Test_set Variables')
print(y_test.value_counts()/test_cnt)
```

```python
# Use roc_auc_score(same evaluation method with Kaggle)

from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# random_state = 42 to derive constant results
xgb_clf = XGBClassifier(n_estimators=100, random_state=42)

xgb_clf.fit(X_train, y_train, early_stopping_rounds=100,
            eval_metric="auc", eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],
                              average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

-> ROC AUC: 0.8378



```python
# Finding the best working parameter for GridSearchCV(finding important features)

from sklearn.model_selection import GridSearchCV

xgb_clf = XGBClassifier(n_estimators=100)

params = {'max_depth':[5, 7] , 'min_child_weight':[1, 3] ,'colsample_bytree':[0.5, 0.75] }
 
gridcv = GridSearchCV(xgb_clf, param_grid=params)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric="auc",
           eval_set=[(X_train, y_train), (X_test, y_test)])

print('GridSearchCV Optimal Parameter:',gridcv.best_params_) 

xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

-> GridSearchCV Optimal Parameter: colsample_bytree: 0.75, max_depth: 7, min_child_weight: 3

ROC AUC: 0.8404

```python
# Added the derived parameters, increased n_estimators to 1000

xgb_clf = XGBClassifier(n_estimators=1000, random_state=42, max_depth=7,
                        min_child_weight=3, colsample_bytree=0.75)

xgb_clf.fit(X_train, y_train, early_stopping_rounds=200, 
            eval_metric="auc",eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

## 1. Original Data Set

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score

y = df_train.TARGET
X = df_train.drop(['TARGET', 'ID'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# random_state = 42

clf = DecisionTreeClassifier(random_state = 42) 
dt_1 = clf.fit(X_train, y_train)

dt_1.score(X_train, y_train), dt_1.score(X_test, y_test)
```

```python
pred = dt_1.predict(X_test)
print("Accuracy with Decision Tree = ", accuracy_score(y_test, pred))
```

```python
plt.figure(figsize=(100, 100))
plot_tree(dt_1, fontsize=10, filled=True)
plt.show()
```

```python
# Submission file - original

x_test_final = test.drop(['ID'], axis=1)
final_prediction = dt_1.predict(x_test_final)

submission_1 = pd.DataFrame({
        "ID": test["ID"],
        "TARGET": final_prediction
    })
submission_1.to_csv('/content/drive/My Drive/CIS508/JM1/submission_file/submission_DecisionTree_orignial.csv',header=True, index=False)
```

### 2-2. Balancing Data Set (SMOTE)

## 3. Selecting Parameter Values

### 3-1. Optimal Random_state Value

```python
from sklearn.tree import DecisionTreeClassifier

result = pd.DataFrame(columns=['train score', 'test score'])

for i in range(0, 100):
    tree = DecisionTreeClassifier(random_state=i)
    tree.fit(X_tr, y_tr)
    tree.score(X_test, y_test)
    i+=1
    result = result.append(pd.DataFrame([[tree.score(X_tr, y_tr), tree.score(X_test, y_test), i-1]],
                                        columns=['train score', 'test score', 'i']))
```

```python
result.sort_values(by = 'test score', ascending = False)
```

### 3-2. Optimal Parameter Combination

```python
# With Optimized random_state

from sklearn.model_selection import GridSearchCV

dt_clf = DecisionTreeClassifier(random_state = 57)
parameters = {'criterion': ['gini', 'entropy'],
              'max_depth': [x for x in range(1, 6,2)],
              'min_samples_leaf': [x for x in range(1, 15,2)],
              'splitter': ['best', 'random']}

grid_dt = GridSearchCV(dt_clf, param_grid = parameters)

grid_dt.fit(X_tr, y_tr)

result = pd.DataFrame(grid_dt.cv_results_['params'])
result['mean_test_score'] = grid_dt.cv_results_['mean_test_score']
result.sort_values(by='mean_test_score', ascending=False)
```

-> criterion = gini, max_depth = 5, min_samples_leaf =	11, splitter =	best
